```python
import numpy as np
```

```python
# converting ratings into a float ranging between 0-1 (will be treated as the targe)

y = np.divide(df['rating'], 5.0)
x = df.loc[:, ['text']]
```

```python
import sklearn
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.2, random_state=42)
```

```python
from transformers import AutoTokenizer, RobertaForSequenceClassification
```

```python
# starting with 5 labels, since there are 5 diff values of 

model = RobertaForSequenceClassification.from_pretrained('FacebookAI/roberta-base', problem_type='multi_label_classification', num_labels=5)
tokenizer = AutoTokenizer.from_pretrained('FacebookAI/roberta-base')
```

```python
# getting rid of punctuation

x_train['text'] = x_train['text'].str.replace(r'[^\w\s]+', '', regex=True)
```

```python
x_train_updated = [item[0] for item in x_train.values.tolist()]
```

```python
input = tokenizer(x_train_updated, return_tensors='pt')
```

```python
import torch
```

```python
with torch.no_grad():
    logits = model(**input).logits
```
