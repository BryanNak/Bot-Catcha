```python
import pandas as pd
import sklearn
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import RidgeCV
from joblib import dump
from joblib import load

df = pd.read_parquet("hf://datasets/theArijitDas/Fake-Reviews-Dataset/data/train-00000-of-00001.parquet")
```

```python
df.count()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
x=df.loc[:, ~df.columns.isin(['label', 'category'])]
y=df['label']
tfidf_vectorizer = TfidfVectorizer()
vectorized_reviews = tfidf_vectorizer.fit_transform(x['text'])
x = vectorized_reviews
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.2, random_state=42, shuffle=True)
```

```python
x_train.shape
```

```python
random_forest = sklearn.ensemble.RandomForestClassifier()
param_grid = {'n_estimators': [100, 300, 1000], 'max_features': ['sqrt', 'log2', .2, .5, 1], 'max_depth': [None, 10, 20, 40], 'min_samples_leaf':[1, 2, 5, 10], 'min_samples_split':[2,5,10]}
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
rand_search_cv = RandomizedSearchCV(random_forest, param_grid, n_jobs=-1, scoring=['accuracy','precision', 'recall', 'f1', 'roc_auc'], refit='recall')
rand_search_cv.fit(x_train, y_train)
```

```python
ridge_reg = sklearn.linear_model.RidgeCV(store_cv_results=True).fit(x_train, y_train)
```

# saving model
dump(ridge_reg, 'ridge_reg.joblib')
